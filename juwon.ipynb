Data Cleaning.

```python
label = "age_of_respondent"

#Drop rows with no data
sam_data = sam_data.dropna(subset = [label])

#drop rows with extreme outliers
sam_data = sam_data.drop(sam_data[sam_data[label] > 18].index)

#checking if the number make sense
sam_data[[label]].describe()
```

```python

label_data = sam_data[label]
fig = plt.figure(figsize=(4, 2))
ax = fig.gca()
ax.hist(label_data, bins=100)
ax.set_ylabel("Frequency")
ax.axvline(label_data.mean(), color="yellow", linestyle = "dashed", linewidth = 2)
ax.axvline(label_data.median(), color="green", linestyle = "dashed", linewidth = 2)
```
